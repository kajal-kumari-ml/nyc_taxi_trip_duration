```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
#importing Linear Regression and metric mean square error
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error as mae
```

```python
data = pd.read_csv('nyc_taxi_trip_duration.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'])
data['dropoff_datetime'] = pd.to_datetime(data['dropoff_datetime'])
```

```python
data['month'] = data.pickup_datetime.dt.month
data['weekday_num'] = data.pickup_datetime.dt.weekday
data['pickup_hour'] = data.pickup_datetime.dt.hour
```

```python
data.head()
```

```python
data['distance']=(np.abs(data['pickup_longitude']))-(np.abs(data['dropoff_latitude']))+(np.abs(data['pickup_longitude']))-(np.abs(data['dropoff_latitude']))
data['speed']=(data.distance/data.trip_duration/3600)
```

```python
data.speed
```

```python
data.dtypes.reset_index()
```

```python
dummy = pd.get_dummies(data.month, prefix='month')
dummy.drop(dummy.columns[0], axis=1, inplace=True) #avoid dummy trap
data = pd.concat([data,dummy], axis = 1)
```

```python
dummy = pd.get_dummies(data.vendor_id, prefix='vendor_id')
dummy.drop(dummy.columns[0], axis=1, inplace=True) #avoid dummy trap
data = pd.concat([data,dummy], axis = 1)
```

```python
dummy = pd.get_dummies(data.weekday_num, prefix='weekday_num')
dummy.drop(dummy.columns[0], axis=1, inplace=True) #avoid dummy trap
data = pd.concat([data,dummy], axis = 1)
```

```python
dummy = pd.get_dummies(data.pickup_hour, prefix='pickup_hour')
dummy.drop(dummy.columns[0], axis=1, inplace=True) #avoid dummy trap
data = pd.concat([data,dummy], axis = 1)
```

```python
dummy = pd.get_dummies(data.passenger_count, prefix='passenger_count')
dummy.drop(dummy.columns[0], axis=1, inplace=True) #avoid dummy trap
data = pd.concat([data,dummy], axis = 1)
```

```python
dummy = pd.get_dummies(data.store_and_fwd_flag, prefix='flag')
dummy.drop(dummy.columns[0], axis=1, inplace=True) #avoid dummy trap
data = pd.concat([data,dummy], axis = 1)
```

```python
y = data.iloc[:,10].values
x = data.iloc[:,range(14,59)].values
```

```python
x.shape,y.shape
```

```python
# Importing the train test split function
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y, random_state = 56)
```

```python
# Creating instance of Linear Regresssion
lr = LR()

# Fitting the model
lr.fit(train_x, train_y)
```

```python
train_predict = lr.predict(train_x)
k = np.sqrt(mae(train_predict, train_y))
print('Training Mean square root Error', k )
```
